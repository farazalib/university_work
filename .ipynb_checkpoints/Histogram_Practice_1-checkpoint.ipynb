We'll continue working with the Pokémon dataset in this workspace.

```python
pokemon = pd.read_csv('pokemon.csv')
pokemon.head()
```

**Task**: Pokémon have a number of different statistics that describe their combat capabilities. Here, create a _histogram_ that depicts the distribution of 'special-defense' values taken. **Hint**: Try playing around with different bin width sizes to see what best depicts the data.

```python
#code here
# Create a histogram using Matplotlib
plt.figure(figsize=(8, 6))  # Adjust the figure size as needed
plt.hist(pokemon['special-defense'], bins=20, edgecolor='black', color='skyblue')

# Customize the plot
plt.title('Distribution of Special Defense in Pokémon')
plt.xlabel('Special Defense Value')
plt.ylabel('Number of Pokémon')
plt.grid(True)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data=[
      ['country','gold_medal'],
      ['united_States',46.0],
      ['Great_britain',27.0],
      ['china',26.0],
      ['Russia',19.0],
      ['Germany',17.0]
                     ]
    
df=pd.DataFrame(data,columns=['country','Gold_medal'])
df.to_csv('Medal.csv',index=False)
                     
medal=pd.read_csv('Medal.csv')
medal
```

```python
medal.dtypes
```

```python
df['Gold_medal'] = pd.to_numeric(df['Gold_medal'], errors='coerce')
# Sort the DataFrame by gold medal achievements and select the top 5 countries
top_countries = df.nlargest(5, 'Gold_medal')


                   
```

```python
# Create a pie chart
plt.figure(figsize=(8,8))
plt.pie(top_countries['Gold_medal'], labels=top_countries['country'], autopct='%1.1f%%', startangle=140)
plt.title('Gold Medal Achievements in 2016 Summer Olympics (Top 5 Countries)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the pie chart
plt.show()
```
